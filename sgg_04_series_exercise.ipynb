```python
# 1. 学生成绩统计
# 创建一个包含10名学生数学成绩的Series，成绩范围在50-100。
# ● 计算平均分、最高分、最低分
# ● 找出高于平均分的学生人数
import pandas as pd
import numpy as np

np.random.seed(42)
arr = np.random.randint(50, 101, 10)
print(arr)
s = pd.Series(arr, index=['学生' + str(i) for i in range(1, 11)], name='score')
print(s.mean())
print(s.max())
print(s.min())
print(s[s>s.mean()].count())
print(s)
print(s.values)
```

```python
# 2. 温度数据分析
# 给定某城市一周每天的最高温度Series，完成以下任务：
# ● 找出温度超过30度的天数
# ● 计算平均温度
# ● 将温度从高到低排序
# ● 找出温度变化最大的两天
temps = pd.Series([28, 31, 29, 32, 30, 27, 33], name='Temperature', index=['星期' + str(i) for i in range(1, 8)])
print(temps[temps>30].count())
print(temps.mean())
print(temps.sort_values(ascending=False)) # 并不改原来的值

diff = pd.Series([np.nan for _ in range(6)])
for i in range(len(temps) - 1):
    diff.iloc[i] = (temps.iloc[i + 1] - temps.iloc[i])
print(diff.abs().argmax())

# 内置方法(很常用的方法)
print(temps.diff().abs().sort_values(ascending=False).keys()[:2].tolist())
# 或
print(temps.diff().abs().sort_values(ascending=False).head(2).keys().tolist())
```

```python
# 3. 给定某股票连续10个交易日的收盘价Series：
# ● 计算每日收益率（当日收盘价/前日收盘价-1）
# ● 找出收益率最高和最低的日期
# ● 计算波动率（收益率的标准差）
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2], name='Price', index=pd.date_range('2024-01-01', periods=10))

# 下面方法不行，我故意是因为他是按照日期除的
print(prices)
prices_diff = prices.diff()
print(prices_diff.tail(9))
print(prices.head(9))
rate = prices_diff.tail(9) / prices.head(9)
print(rate)
print(rate.argmax())
print(rate.std())


# 内置方法
a = prices.pct_change()
print(a)
print(a.idxmax()) # 最大值对应的标签，与argmax的区别：一个返回索引，一个返回标签
print(a.std())
```

```python
# # 4. 某产品过去12个月的销售量Series：
# ● 计算季度平均销量
# ● 找出销量最高的月份
# ● 计算月环比增长率
# ● 找出连续增长超2个月的月份
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220], index=pd.date_range('2022-01-01', periods=12, freq='MS'))
print(sales)
print(sales.resample('QS').mean()) # .resample的索引类型必须是datatime类型的
print(sales.idxmax())
a = sales.pct_change()
print(a)
b = a>0
print(b[b.rolling(3).sum() == 3].keys())

```

```python
# 5. 每小时销售数据分析
# 某商店每小时销售额Series：
# ● 按天重采样计算每日总销售额
# ● 计算每天营业时间(8:00-22:00)和非营业时间的销售额比例
# ● 找出销售额最高的3个小时
np.random.seed(42)
sales = pd.Series(np.random.randint(0, 100, 24), index=pd.date_range('2025-01-01', periods=24, freq='h'))
print(sales)
print(sales.resample('D').sum())
print(sales.between_time('8:00', '22:00'))

# 或者bool索引
print(sales[(sales.index.hour>=8) & (sales.index.hour<=22)])
# 或iloc切片
print(sales.iloc[8:23])

print(sales.sort_values().head(3).keys())
```
