## 1. 折线图：展示趋势

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 设置字体
rcParams['font.family'] = 'SimHei'

# 0 准备数据
month = ['1月', '2月', '3月', '4月']
sales = [100, 150, 80, 130]

# 1. 创建画布，设置大小
plt.figure(figsize=(10, 5))

# 2. 绘制折线图(实际上只要设置这个就行，其他的不用改)
# 别的类型的图也基本上只需要改这个就行
plt.plot(month, sales, 
         label='产品A', # legend就是设置这个
         color='orange',
         linewidth=2,
         linestyle='--',
         marker='o' # 数据点用圆点显示
         )

# 3. 设置
plt.title('2025年销售趋势', color='red', fontsize=20) # 标题
plt.xlabel('月份', fontsize=10) # x轴标签
plt.ylabel('销售额(万元)', fontsize=10) # y轴标签
plt.legend(loc='upper left') # 图例

# 网格先：True: xy都有；axit='x'，则只有垂直；axis='y'，则只有水平
# plt.grid(axis='y')
# plt.grid(axis='x')
plt.grid(True, alpha=0.1, color='blue', linestyle='--')

# 设置刻度
plt.xticks(rotation=0, fontsize=10) # 如果刻度比较密集的话，rotate
plt.yticks(rotation=0, fontsize=10)

# 设置y轴范围
plt.ylim(0, 160)
plt.xlim(0, )

# 每个点标上值
for x, y in zip(month, sales):
    # print(x, y)
    # x, y, str(y) 分别为x坐标，y坐标，标注的值
    plt.text(x, y+2, str(y), ha='center', va='bottom', fontsize=10) # ha horizon align

# 显示图表
plt.show()
```

# 2. 柱状图(纵向条形图)

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 设置字体
rcParams['font.family'] = 'SimHei'

# 0 准备数据
subject = ['语文', '数学', '英语', '科学']
scores = [85, 92, 78, 88]

# 1. 创建画布，设置大小
plt.figure(figsize=(10, 5))

# 2. 绘制条形图(实际上只要设置这个就行，其他的不用改)
# 别的类型的图也基本上只需要改这个就行
plt.bar(subject, scores, 
         label='小红', # legend就是设置这个
         color='blue',
         width=0.6,
         )

# 3. 设置
plt.title('2025成绩', color='red', fontsize=20) # 标题
plt.xlabel('科目', fontsize=10) # x轴标签
plt.ylabel('分数', fontsize=10) # y轴标签
plt.legend(loc='upper left') # 图例

# 网格线：True: xy都有；axit='x'，则只有垂直；axis='y'，则只有水平
plt.grid(axis='y', alpha=0.1, color='blue', linestyle='--')
# plt.grid(axis='x')
# plt.grid(True, alpha=0.1, color='blue', linestyle='--')

# 设置刻度
plt.xticks(rotation=0, fontsize=10) # 如果刻度比较密集的话，rotate
plt.yticks(rotation=0, fontsize=10)

# 设置y轴范围
plt.ylim(0, 100)
# plt.xlim(0, )

# 每个点标上值
for x, y in zip(subject, scores):
    # print(x, y)
    # x, y, str(y) 分别为x坐标，y坐标，标注的值
    plt.text(x, y, str(y), ha='center', va='bottom', fontsize=10) # ha horizon align

# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

# 3. 条形图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 设置字体
rcParams['font.family'] = 'SimHei'

# 0 准备数据
countries = ['中国', '美国', '俄罗斯', '意大利']
gdp = [85, 92, 78, 88]

# 1. 创建画布，设置大小
plt.figure(figsize=(10, 5))

# 2. 绘制条形图(实际上只要设置这个就行，其他的不用改)
# 别的类型的图也基本上只需要改这个就行
plt.barh(countries, gdp, 
         label='', # legend就是设置这个
         color='blue',
         height=0.6,
         )

# 3. 设置
plt.title('GDP排名', color='red', fontsize=20) # 标题
plt.ylabel('国家', fontsize=10) # x轴标签
plt.xlabel('GDP', fontsize=10) # y轴标签
plt.legend(loc='upper left') # 图例

# 网格线：True: xy都有；axit='x'，则只有垂直；axis='y'，则只有水平
# plt.grid(axis='y', alpha=0.1, color='blue', linestyle='--')
plt.grid(axis='x', alpha=0.1, color='blue', linestyle='--')
# plt.grid(True, alpha=0.1, color='blue', linestyle='--')

# 设置刻度
plt.xticks(rotation=0, fontsize=10) # 如果刻度比较密集的话，rotate
plt.yticks(rotation=0, fontsize=10)

# 设置y轴范围
plt.xlim(0, 100)
# plt.ylim(0, 100)

# 每个点标上值
for x, y in zip(countries, gdp):
    # print(x, y)
    # x, y, str(y) 分别为x坐标，y坐标，标注的值
    plt.text(y+1, x, str(y), ha='center', va='bottom', fontsize=10) # ha horizon align

# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

# 4. 饼图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 设置字体
rcParams['font.family'] = 'SimHei'

# 0 准备数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
time = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']

# 1. 创建画布，设置大小
plt.figure(figsize=(10, 5))

# 2. 饼图设置比较少
plt.pie(time, labels=things, 
        autopct='%.1f%%', # 第一个%是固定格式，表示取后面的格式，第二个是转义%
        startangle=90,
        colors=colors
        )

# 3. 设置
plt.title('时间分配', color='red', fontsize=20) # 标题

# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

# 5. 环形图（与饼图差不多）

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 设置字体
rcParams['font.family'] = 'SimHei'

# 0 准备数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
time = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']

# 1. 创建画布，设置大小
plt.figure(figsize=(10, 5))

# 2. 饼图设置比较少
plt.pie(time, labels=things, 
        autopct='%.1f%%', # 第一个%是固定格式，表示取后面的格式，第二个是转义%
        startangle=90,
        colors=colors,
        wedgeprops={'width': 0.6}, # 环形图专属
        pctdistance=0.65 # 调整百分比位置
        )

# 3. 设置
plt.title('时间分配', color='red', fontsize=20) # 标题
plt.text(0, 0, '总计\n100%', ha='center', va='center', fontsize=10)
# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

# 6. 爆炸饼图(突出中心)

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams # 设置字体
rcParams['font.family'] = 'SimHei'

# 0 准备数据
things = ['学习', '娱乐', '运动', '睡觉', '其他']
time = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
explode = [0.1, 0, 0.2, 0, 0] # 设置突出块的位置

# 1. 创建画布，设置大小
plt.figure(figsize=(10, 5))

# 2. 饼图设置比较少
plt.pie(time, labels=things, 
        autopct='%.1f%%', # 第一个%是固定格式，表示取后面的格式，第二个是转义%
        # startangle=90, # 一般右上角是人眼注意到的地方
        colors=colors,
        explode=explode,
        shadow=True # 加阴影
        )

# 3. 设置
plt.title('时间分配', color='red', fontsize=20) # 标题

# 自动优化排版
plt.tight_layout()
# 显示图表
plt.show()
```

# 7. 散点图

```python
import matplotlib.pyplot as plt
import random
from matplotlib import rcParams # 设置字体
rcParams['font.family'] = 'SimHei'

# 0 准备数据
x = []
y = []
for i in range(1000):
    temp = random.uniform(0, 10)
    x.append(temp)
    temp2 = temp * 2 + random.gauss(0, 3)
    y.append(temp2)

# 1. 创建画布，设置大小
plt.figure(figsize=(10, 5))

# 2. 绘制散点图(实际上只要设置这个就行，其他的不用改)
# 别的类型的图也基本上只需要改这个就行
plt.scatter(x, y, 
            color='blue',
            alpha=0.6, # 重复的可以重叠
            s=20, # 设置圆点大小
            label='数据'
            )
# 回归线
plt.plot([0, 10], [0, 20], color='red', linewidth=2)

# 3. 设置
plt.title('x与y的关系', color='red', fontsize=20) # 标题
plt.xlabel('x变量', fontsize=10) # x轴标签
plt.ylabel('y变量', fontsize=10) # y轴标签
plt.legend(loc='upper left') # 图例

# 网格先：True: xy都有；axit='x'，则只有垂直；axis='y'，则只有水平
# plt.grid(axis='y')
# plt.grid(axis='x')
plt.grid(True, alpha=0.1, color='blue', linestyle='--')

# 设置刻度
plt.xticks(rotation=0, fontsize=10) # 如果刻度比较密集的话，rotate
plt.yticks(rotation=0, fontsize=10)

# 设置y轴范围
plt.ylim(0, 30)
plt.xlim(0, 10)

# 每个点标上值
# for x, y in zip(x, y):
#     # print(x, y)
#     # x, y, str(y) 分别为x坐标，y坐标，标注的值
#     plt.text(x, y+2, str(y), ha='center', va='bottom', fontsize=10) # ha horizon align

# 显示图表
plt.show()
```

# 8. 箱线图

# 多图绘制

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'

month = ['1月','2月','3月','4月']
sales = [100, 150, 80, 130]

# 动态图表的生成，没有事先定义有多少个图
f1 = plt.subplot(2, 2, 1) # 2 * 2 个表格的第1张图
f1.plot(month, sales)
f2 = plt.subplot(2, 2, 2)
f2.bar(month, sales)
f3 = plt.subplot(2, 2, 3)
f3.scatter(month, sales)
f4 = plt.subplot(2, 2, 4)
f4.barh(month, sales)

plt.show()
```

# 案例

```python
# 天气分析 csv
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'SimHei'

# 读取数据
df = pd.read_csv('./data/weather.csv')
print(df.head())

# 数据预处理
print(df.isna().sum()) # 没有缺失值，很好
# date变为datatime类型
df['date'] = pd.to_datetime(df['date'])
# print(df.info())
# 好多年的，太多了
df = df[df['date'].dt.year == 2015]

plt.figure(figsize=(20, 10))
# 最高气温
plt.plot(df['date'], df['temp_max'], label='最高')
# 最低气温
plt.plot(df['date'], df['temp_min'],label='最低')
# 平均
plt.plot(df['date'], (df['temp_max'] + df['temp_min']) / 2, label='平均')

plt.title('2015年气温变化', fontsize=10)
plt.ylabel('气温')
plt.xlabel('时间')
plt.legend() # 图例默认右上角

plt.show()
```
