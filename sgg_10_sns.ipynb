# seaborn
## matplotlib的封装

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'SimHei'
penguins = pd.read_csv('./data/penguins.csv')
# print(penguins.head())

# 数据清洗
# print(penguins.isna().sum())
penguins.dropna(inplace=True)
# print(penguins.isna().sum())

```

```python
# 1. 直方图
sns.histplot(data=penguins, x='species') # data是数据集，x=取数据。而matplotlib要手动取数据
```

```python
# 2. KDE图
sns.kdeplot(data=penguins, x='bill_length_mm')
```

```python
# 3. 直方图同时显示kde
sns.histplot(data=penguins, x='bill_length_mm', kde=True)
```

```python
# 4. 计数图(与直方图很像，连续区间只能用直方图了)
sns.countplot(data=penguins, x='island')
```

```python
# 5. 散点图
sns.scatterplot(data=penguins, x='body_mass_g', y='flipper_length_mm', hue='sex')
```

```python
# 6. 蜂窝图
sns.jointplot(data=penguins, x='body_mass_g', y='flipper_length_mm', kind='hex')
```

```python
# 二维核密度
sns.kdeplot(data=penguins, x='body_mass_g', y='flipper_length_mm') # 两个变量的
```

```python
# 填充颜色的kdeplot
sns.kdeplot(data=penguins, x='body_mass_g', y='flipper_length_mm', fill=True, cbar=True)
```

```python
# 条形图
sns.barplot(data=penguins, x='species', y='bill_length_mm', estimator='mean', errorbar=None)
```

```python
# 箱线图
sns.boxplot(data=penguins, x='species', y='bill_length_mm')
```

```python
# 小提琴图 = 箱线图 + KDE
# 小提琴图（Violin Plot）是一种结合了箱线图和核密度估计图（KDE）的可视化图表，
# 用于展示数据的分布情况、集中趋势、散布情况以及异常值。
# 小提琴图不仅可以显示数据的基本统计量（如中位数和四分位数），
# 还可以展示数据的概率密度，提供比箱线图更丰富的信息
sns.violinplot(data=penguins, x='species', y='bill_length_mm')
```

```python
# 成对关系图
# 成对关系图是一种用于显示多个变量之间关系的可视化工具。
# 它可以展示各个变量之间的成对关系，
# 并且通过不同的图表形式帮助我们理解数据中各个变量之间的相互作用。

# 对角线上的图通常显示每个变量的分布（如直方图或核密度估计图），帮助观察每个变量的单变量特性。
# 其他位置展示所有变量的两两关系，用散点图表示。

# 不知道什么关系的情况下，这样一画就知道了

# 对角线默认为kde，其他为散点图
# hue为根据什么图颜色
sns.pairplot(data=penguins, hue='species', diag_kind='kde', kind='scatter')
```
